```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Dataset

```python
X = [[27],
     [35],
     [39],
     [21],
     [22],
     [45]]

y_names = ['lemon', 'orange', 'orange', 'lemon', 'lemon', 'orange']
y = [0, 1, 1, 0, 0, 1]
```

```python
df = pd.DataFrame(X, columns=['size'], index=y_names,) # Column name changed to 'size'
df
df['y_names'] = y_names
df['y'] = y
df
```

## Data Visualization

```python
# plt.figure(figsize=(4,3))

plt.xlabel('fruit')
plt.ylabel('size')
plt.scatter(y_names, X)
plt.show()
```

```python
X_input_size = 55

if (X_input_size <= 31):
    fruit = 'Lemon'
else:
    fruit = 'Orange'

print(fruit)
```

# Model
https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
model.fit(X, y)
```

## Predict

```python
X_input = [[20.3]]
y_predict = model.predict(X_input)
y_predict
# print('Result =', y_predict[0] )
```

```python
fruit_classes = ['Lemon','Orange']
```

```python
prd = fruit_classes[y_predict[0]]
print('Result: {}'. format(prd))
```

```python
X_input = [[31.0],
           [25.0],
           [21.2],
           [38.6]]
y_predict = model.predict(X_input)
y_predict
```

```python
print('Result:')
for i in y_predict:
    print(fruit_classes[i])
```

```python
for i, val in enumerate(y_predict):
    print(f'{X_input[i]} = {val}  {fruit_classes[val]}')
```

# Classification 2: (Fruit 3 classes)
Lemon, Orange and Pomelo

## Dataset

```python
X = [[27],
     [35],
     [39],
     [21],
     [22],
     [45],
     [102],
     [110],
     [120] ]

y_names = ['lemon', 'orange', 'orange', 'lemon', 'lemon', 'orange', 'pamelo', 'pamelo', 'pamelo']
y = [0, 1, 1, 0, 0, 1, 2, 2, 2]
```

## Data Visualization

```python
plt.xlabel('fruit')
plt.ylabel('size')

plt.scatter(y_names, X)
plt.show()
```

## Model

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
fruit_classes = ['Lemon','Orange', 'Pamelo']
```

## Predict

```python
X_input = [[23]]
y_predict = model.predict(X_input)

prd = fruit_classes[y_predict[0]]
print('Result : {}'. format(prd))
```

```python
X_input = [[28],
     [32],
     [39],
     [101]]

y_predict = model.predict(X_input)

y_predict
# class_names[y_predict]
# prd = class_names[y_predict[0]]
# print('Result : {}'. format(prd))
```

```python
for i, k in enumerate(y_predict):
    print(f'{X_input[i]} = {fruit_classes[k]}')
```

## Visualize Model

### Tree

```python
from sklearn.tree import plot_tree

data_feature_names = ['size']

plt.subplots(nrows=1, ncols=1, figsize=(5, 6), dpi=96)
# plt.figure(figsize=(6,6))

plot_tree(model,
         filled=True, rounded=True,
         feature_names=data_feature_names,
         class_names=fruit_classes)

plt.show()
```

```python
X_input_size = 72.4

if (X_input_size <= 31):
    fruit = 'Lemon'
elif (X_input_size <= 73.5):
    fruit = 'Orange'
else:
    fruit = 'Pomelo'

print(fruit)
```
