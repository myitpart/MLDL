# Machine Learning

```python
x = np.array(x_data)
y = np.array(y_data)

x
```

```python
# ws02
x = x.reshape(-1, 1)
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
model.predict([[2]])
```

```python
x_input = [[2.0],
           [2.5],
           [3],
           [5.0]]

# x_input = [[6.0]]

y_predict = model.predict(x_input)
# y_predict[1]
# y_predict[2]
y_predict
```

```python
x_input = [2.0, 2.5, 3, 5.0]
x_input = np.array(x_input).reshape(-1, 1)
x_input
```

```python
y_predict = model.predict(x_input)
y_predict
```

```python
y2_predict = y_predict.reshape(1, -1)

x_input_list = x_input
# print(x_input_list)

for i, item in enumerate(x_input_list):
#     print(type(k))
    print('x={} y={:.5f} '.format(item[0], y2_predict[0][i]) )
```

```python
prd = model.predict(x)

plt.rcParams['figure.figsize'] = 5,3

plt.grid()
plt.scatter(x, y, color='b')
plt.plot(x, prd, linewidth='1')
plt.scatter(x, prd, color='r', marker='s')
plt.show()
```

```python
model.coef_ , model.intercept_  # m and b
```

```python
# ws03

import joblib

joblib.dump(model, 'my_model.pkl')
```
